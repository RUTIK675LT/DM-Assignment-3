```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/ecommerce crisp dm dataset.csv')

# Display the first few rows of the dataset
df.head()
```

```python
# Basic statistics of the dataset
stats = df.describe()

# Check for missing values
missing_values = df.isnull().sum()

stats, missing_values
```

```python
# Encode the 'Customer Gender' column
df_encoded = pd.get_dummies(df, columns=['Customer Gender'], drop_first=True)

# Display the encoded dataframe
df_encoded.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the scaler
scaler = MinMaxScaler()

# Scale the 'Customer Shopping Frequency' column
df_encoded['Customer Shopping Frequency'] = scaler.fit_transform(df_encoded[['Customer Shopping Frequency']])

# Display the scaled dataframe
df_encoded.head()
```

```python
from pycaret.regression import *

# Set up the pycaret environment for regression
setup_config = setup(data = df_encoded, target = 'Customer Shopping Amount', session_id=123)
```

```python
best_model = compare_models()
```

```python
save_model(best_model, 'final_model')
```
